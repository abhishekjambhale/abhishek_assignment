## Problem statement - To find the Balance of the new user by using data of users available in credit.csv

```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split as tts
import warnings
warnings.filterwarnings("ignore")
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import preprocessing
from sklearn.linear_model import Ridge,RidgeCV, ElasticNet,ElasticNetCV, Lasso, LassoCV
```

```python
df = pd.read_csv('Credit.csv')
```

```python
df.head()
```

```python
df = df.drop(['Unnamed: 0'],1)
```

```python
df.dtypes
```

## Applying Linear Regression on Credit.csv file using all variables.

## We have to use one hot encoding here for variables which are objects.

```python
df = pd.get_dummies(df)
df_scale = StandardScaler().fit_transform(df)
```

```python
df.head()
```

```python
X = df.drop(['Balance'],1)
Y = df['Balance']
```

```python
df.dtypes
```

#### Linear Regression

```python
X_train,X_test,Y_train,Y_test = tts(X,Y,test_size=0.25,random_state = 42)
lin_rig = LinearRegression(normalize=True)
lin_rig.fit(X_train,Y_train)
Y_pred = lin_rig.predict(X_test)
print('r2 Score of LinearRegression model is : ', r2_score(Y_test,Y_pred))
```

#### Ridge Regression with cross validation

```python
reg = RidgeCV()
reg.fit(X,Y)
print("Best alpha using build in RidgeCV: %f" % reg.alpha_)
print(reg.score(X,Y))
```

#### Lasso Regression with CV

```python
reg_1 = LassoCV()
reg_1.fit(X,Y)
print("Best alpha using build in LassoCV: %f" % reg_1.alpha_)
print(reg_1.score(X,Y))
```

#### ElasticNet CV

```python
reg_2 = ElasticNetCV()
reg_2.fit(X,Y)
print("Best alpha using build in ElasticnetCV: %f" % reg_2.alpha_)
print(reg_2.score(X,Y))
```

### Dropping all other variables

```python
df = pd.read_csv('Credit.csv')
```

```python
df = df.drop(['Unnamed: 0','Cards','Age','Education','Gender','Student','Married','Ethnicity'],1)
```

```python
X1 = df.drop(['Balance'],1)
Y1 = df['Balance']
```

#### Linear Regression

```python
X1_train,X1_test,Y1_train,Y1_test = tts(X1,Y1,test_size=0.25,random_state = 42)
lin_rig = LinearRegression(normalize=True)
lin_rig.fit(X1_train,Y1_train)
Y1_pred = lin_rig.predict(X1_test)
print('r2 Score of LinearRegression model is : ', r2_score(Y1_test,Y1_pred))
```

#### Ridge Regression with cross validation

```python
reg = RidgeCV()
reg.fit(X1,Y1)
print("Best alpha using build in RidgeCV: %f" % reg.alpha_)
print(reg.score(X1,Y1))
```

#### Lasso Regression with CV

```python
reg_1 = LassoCV()
reg_1.fit(X1,Y1)
print("Best alpha using build in LassoCV: %f" % reg_1.alpha_)
print(reg_1.score(X1,Y1))
```

#### ElasticNet CV

```python
reg_2 = ElasticNetCV()
reg_2.fit(X1,Y1)
print("Best alpha using build in ElasticnetCV: %f" % reg_2.alpha_)
print(reg_2.score(X1,Y1))
```
